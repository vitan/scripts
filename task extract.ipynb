```python
import re

fls = open('/home/weizhou/Projects/scripts/input/service/fls', 'r')
cr = open('/home/weizhou/Projects/scripts/input/service/cr', 'r')
lb_service = open('/home/weizhou/Projects/scripts/input/service/lb', 'r')
lb_task = open('/home/weizhou/Projects/scripts/input/task/lb', 'r')
service = open('/home/weizhou/Projects/scripts/output/service.raw', 'r')

outf= open('/home/weizhou/Projects/scripts/output/task', 'w')
```

```python
#
# Get all the tasks
#

service_code_reg = re.compile(r'(FS|CR|LB)-[\d]{3}-[\d]{2}-[\d]{3}-v[\d]{1}')

#1	Engineering Operations - CORE
#2	Engineering Operations - LAB
#3	Engineering Operations - FLS
team_dict = {'CR': '1',
             'LB': '2',
             'FS': '3',
             'CA': '4',
             'SA': '5',
             'SE': '6',
             }

service_dict = dict()
for item in service:
    item = item.strip()
    for element in item.split('\t')[1].split(' | '):
        service_dict[element] = item.split('\t')[0]

current_service_id = None
task_dict = dict()

# FLS task process
for line in fls:
    row_list = line.split('\t')
    #FLS table format
    #Service Name \t Service Code \t Tasks Performed \t Link to Documentation
    #As some boring dirty data, cleaned the data manully in advance.
    if len(row_list) is not 4:
        raise Exception('ERROR FORMAT', line)
    if row_list[0].strip():
        current_service_id = service_dict[row_list[0]]
        
    code = row_list[1].strip()
    link = '[KB:'+row_list[3].strip()
    task_dict[code if code else'[NA]'+row_list[2]] = [current_service_id,
                                      team_dict[code.split('-')[0]] if code else'NA',
                                      code.split('-')[3] if code else'0',
                                      row_list[2].strip() if row_list[2].strip() else 'NA',
                                      link.replace(' (', '] (') if link.find(' (') != -1 else link + ']']

#lb task process
for line in lb_service:
    row_list = line.split('\t')
    #LB table format
    #Service ID \t Service Name
    if len(row_list) is not 2:
        raise Exception('ERROR FORMAT', line)
    service_dict[row_list[0].split('-')[1]] = service_dict[row_list[1].strip()]

for line in lb_task:
    if service_code_reg.match(line):
        row_list = line.split('\t')
        if len(row_list) is not 6:
            raise Exception('ERROR FORMAT', line)
        code = row_list[0].strip()
        link = '[KB:'+row_list[5].strip()
        task_dict[code] = [service_dict[code.split('-')[1]],
                           team_dict[code.split('-')[0]],
                           code.split('-')[3],
                           row_list[1].strip() if row_list[1].strip() else 'NA',
                           link.replace(' (', '] (') if link.find(' (') != -1 else link + ']'
                           ]

# clean up
NO_MATCHING = 'NO MATCHING TASK AVAILABLE'
result = [item for item in task_dict.values() if item[3].strip() != NO_MATCHING]
count = xrange(0, len(result))
id_task = ["|%s|%s|" % (key+1, '|'.join(value)) for (key, value) in zip(count, result)]
outf.write("||ID||Service ID||Responsible Team ID||Cost||Description||Detail Link||\n" + str('\n'.join(id_task)))
```

```python
fls.close()
cr.close()
lb_service.close()
lb_task.close()
service.close()
outf.close()
```
